```python
import geopandas as gpd
import folium
```

```python
routes = gpd.read_file("../data/navigation.geojson")
pts = gpd.read_file("../data/navigation_points.geojson")
```

```python
endpoint = pts[pts["Type"]=="Endpoint"]
```

```python
bus = pts[pts["Icon"]=="bus"]
ferry = pts[pts["Icon"]=="ferry"]
bike = pts[pts["Icon"]=="person-biking"]
hike = pts[pts["Icon"]=="person-hiking"]
```

```python
m =routes.explore(column="Mode", tooltip="Display", tiles="CartoDB positron", cmap=["purple", "orange", "blue", "red"])
pts.explore(m=m, color="white", tooltip=False, marker_kwds={"stroke":0, "fill":True, "radius":5}, style_kwds={"fillOpacity":1})
endpoint.explore(m=m, color="black", tooltip="Label",
                 style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-location-dot" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),
                 marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
bus.explore(m=m, color="orange", tooltip=False,
                 style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-bus" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),
                 marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
ferry.explore(m=m, color="blue", tooltip=False,
                 style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-ferry" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),
                 marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
bike.explore(m=m, color="purple", tooltip=False,
                 style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-person-biking" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),
                 marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
hike.explore(m=m, color="red", tooltip=False,
                 style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-person-hiking" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),
                 marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
```
