```python
import geopandas as gpd
import folium
```

```python
countries = gpd.read_file("../../day03/data/World_Countries_Generalized.geojson")
```

```python
inland_water = gpd.read_file("../data/DIVA-GIS/JPN_water_lines_dcw.geojson")
```

```python
boat_rentals = gpd.read_file("../data/OSM/boat_rentals_japan.geojson")
```

```python
japan = countries[countries["COUNTRY"] == "Japan"]
```

```python
boat_rentals_japan = gpd.clip(boat_rentals, japan)
```

```python
boat_rentals_japan = boat_rentals_japan.assign(Label='Boat rental')
```

```python
boat_rentals_japan.to_file("../data/OSM/boat_rentals_japan_clip.geojson")
```

```python
m = japan.explore(tiles="CartoDB positron", color="gray", style_kwds={"fill":False, "weight":1, "color":"gray"}, tooltip=False)
inland_water.explore(m=m, style_kwds={"fill":False, "weight":0.5, "color":"#99cfe0"}, tooltip=False)
boat_rentals_japan.explore(m=m, tooltip=False,
                     color="#FFFFF0",
                     style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-sailboat" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:15px"></span>"""},
                                     ),marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
boat_rentals_japan.explore(m=m, tooltip="name", column="Label", legend=True,
                     color="#00008B",
                     style_kwds=dict(style_function=lambda x: {"html": f"""<span   class="fa fa-sailboat" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:14px"></span>"""},
                                     ),marker_type="marker",marker_kwds=dict(icon=folium.DivIcon()))
```
