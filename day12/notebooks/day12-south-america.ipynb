```python
import geopandas as gpd
import pandas as pd
import folium
```

```python
countries = gpd.read_file("../../day03/data/World_Countries_Generalized.geojson")
```

```python
ecuador = countries[countries["COUNTRY"] == "Ecuador"]
```

```python
volcanoes = gpd.read_file("../../ideas/princeton/harvard-glb-volc-geojson.json")
```

```python
volcanoes_ecuador = volcanoes[volcanoes["LOCATION"].isin(["Galapagos", "Ecuador"])]
```

```python
m = ecuador.explore(tiles='CartoDB positron', style_kwds = {"fill": None, "color":"grey"})
volcanoes_ecuador.explore(
    m=m,
    tooltip=["NAME_", "ELEV", "TYPE_", "STATUS"],
    cmap="hot_r",
    column = "ELEV",
    legend_kwds = {"colorbar": True, "max_labels":-2},
    # style_kwds={"style_function":lambda x: {"radius":(x["properties"]["ELEV"]/1000)}},
    style_kwds=dict(
        style_function=lambda x: {
            "html": f"""<span   class="fa fa-volcano" 
                                style="color:{x["properties"]["__folium_color"]};
                                font-size:20px"></span>"""
        },
    ),
    marker_type="marker",
    marker_kwds=dict(icon=folium.DivIcon())
)
m
```

```python
m.save("../outputs/day12.html")
```
