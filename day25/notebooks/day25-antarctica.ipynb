```python
import geopandas as gpd
import pandas as pd
import folium
```

```python
countries = gpd.read_file("../../day03/data/World_Countries_Generalized.geojson")
```

```python
antarctica = countries[countries["COUNTRY"] == "Antarctica"]
```

```python
antarctica_3031 = antarctica.to_crs(epsg=3031)
base = antarctica_3031.plot(color='white', edgecolor='black')
base.set_title("Polar Stereographic")
```
